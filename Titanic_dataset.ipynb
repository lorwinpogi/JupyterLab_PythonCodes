```python
import csv
import matplotlib.pyplot as plt
lines = []
with open('titanic_train.csv', encoding="utf-8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for line in csv_reader:
        lines.append(line)
```

```python
data = lines[1:]
passengers = []
headers = lines[0]
```

```python
for row in data:
    passenger = {headers[i]: row[i] for i in range(len(headers))}
    passengers.append(passenger)
```

```python
survived_counts = [0, 0] 
for p in passengers:
    if p['Survived'] == '1':
        survived_counts[1] += 1
    else:
        survived_counts[0] += 1
```

```python
plt.figure(figsize=(6, 6))
plt.pie(survived_counts, labels=['0', '1'], autopct='%1.1f%%', colors=['lightblue','lightgreen'])
plt.title('Survived')
plt.show()
```

```python
gender_survived = {'female': 0, 'male': 0}
for p in passengers:
    if p['Survived'] == '1':
        gender_survived[p['Sex']] += 1
```

```python
plt.figure(figsize=(6, 6))
plt.bar(gender_survived.keys(), gender_survived.values(), color=['steelblue', 'steelblue'])
plt.title('surviving passengers count by gender')
plt.show()
```
